## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
#Your code here
import sqlite3
import pandas as pd
conn = sqlite3.connect('data.sqlite', detect_types=sqlite3.PARSE_COLNAMES)
cur = conn.cursor()
```

## Display the names of all the employees in Boston.

```python
#Your code here
cur.execute('''SELECT firstname, lastname FROM offices JOIN employees WHERE city = "Boston" ''')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

## Do any offices have no employees?

```python
#Your code here
cur.execute('''SELECT * FROM employees JOIN Offices using(officeCode) ''')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
#print(len(df))
print(len(df[df.officeCode.isnull()]))
#df[df.orderNumber.isnull()].head()
```

## Level Up: Display the names of each product each employee has sold.

```python
# Your code here
cur.execute('''SELECT * FROM employees e
                        JOIN customers c
                        on e.employeeNumber = c.salesRepemployeeNumber 
                        JOIN orders o
                        USING (customerNumber)
                        JOIN orderdetails od
                        USING(orderNumber)
                        JOIN products p
                        USING (productCode)''')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print(len(df))
df

```

## Level Up: Display the Number of Products each Employee Has sold

```python
#Your code here
cur.execute('''SELECT lastName, firstName, quantityOrdered FROM employees e
                        JOIN customers c
                        on e.employeeNumber = c.salesRepemployeeNumber 
                        JOIN orders o
                        USING (customerNumber)
                        JOIN orderdetails od
                        USING(orderNumber)
                        JOIN products p
                        USING (productCode)
                        GROUP BY lastName, firstName''')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```
